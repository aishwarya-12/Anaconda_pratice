Deep Copy - Shallow Copy

```python
marks = np.array([33,44,77,85,45,32,47,88,90])
marks
```

```python
top_marks = marks[:5]          #Shallow copy
top_marks[0]=11       #original marks is also changed
marks
```

```python
top_marks[:] = 99
print(marks)
print(top_marks)
```

```python
marks = np.array([33,44,77,85,45,32,47,88,90])
marks_copy = marks.copy()     #Deep copy
marks_copy[:] = 99
print(marks_copy)
print(marks)          #no Changes in original
```
