```python
from pyspark import SparkContext
```

```python
sc = SparkContext("local","RDDProg").getOrCreate()
```

```python
sc.stop()
```

```python
rdd = sc.parallelize([1,2,3,4])
rdd.collect()
```

```python
rdd.getNumPartitions()
```

```python
rdd = sc.parallelize([1,2,3,4,5,4,3],numSlices=2)
rdd.collect()
```

```python
rdd.getNumPartitions()
```

```python
rdd_squared = rdd.map(lambda x: x**2)
rdd_squared.collect()
```

```python
rdd_filtered = rdd_squared.filter(lambda x: x>10)
rdd_filtered.collect()
```

```python
rdd_filtered = rdd_filtered.distinct()
rdd_filtered.collect()
```

```python
rdd.union(rdd_squared).distinct().collect()
```

```python
rdd.subtract(rdd_squared).collect()
```

```python
rdd_filtered.reduce(lambda a,b : a+b)
```

```python

rdd_filtered.countByValue()
```

**Pair RDD**

```python
my_list = ['Sam 23', 'Mary 34', 'Peter 25','Sam 23', 'Mary 34']
reg_rdd = sc.parallelize(my_list)
pair_rdd = reg_rdd.map(lambda x: (x.split(' ')[0], int(x.split(' ')[1])))
pair_rdd.collect()
```

```python
pair_rdd.reduceByKey(lambda a,b : a+b).collect()
```

```python
pair_rdd_rev = pair_rdd.map(lambda x: (x[1],x[0]))
pair_rdd_rev.sortByKey(ascending=False).collect()
```

```python
pair_rdd_rev.sortBy(lambda x : x,ascending=False).collect()
```

```python
#pair_rdd.groupByKey().mapValues(lambda x: list(x)).collect()
pair_rdd.groupByKey().mapValues(len).collect()
```

```python
pair_rdd.groupByKey().flatMapValues(lambda x: x).collect()
```

```python
pair_rdd.keys().collect()
```

```python
pair_rdd.values().collect()
```

```python
tup = [('a',1),('b',2),('c',3)]
pair_rdd = sc.parallelize(tup)
pair_rdd.collect()
```

```python
pair_rdd1 = pair_rdd.mapValues(lambda x: x+1)
pair_rdd1.collect()
```

```python
pair_rdd.join(pair_rdd1).collect()
```

```python
pair_rdd.countByKey()
```

```python
pair_rdd.countByValue()
```

```python
pair_rdd.count()
```

```python
pair_rdd.lookup('a')
```

**Word Count**

```python
rdd = sc.textFile("file:///content/sample_data/wordcount_input.txt")
rdd.collect()
```

```python
wordCount = rdd.flatMap(lambda x: x.split(' '))\
      .map(lambda x: (x.lower(),1))\
      .reduceByKey(lambda a,b: a+b)
wordCount.collect()
```
