Removing DNB and TDNB matches

```python
df_no_dnb = df[(df['score'] != 'DNB') & (df['score'] != 'TDNB')].copy()
len(df_no_dnb)
```

Converting or Casting to required data types

1.odi_number column

```python
#df_no_dnb['odi_number'] = df_no_dnb['odi_number'].astype('string')
#df_no_dnb['odi_number'] = df_no_dnb['odi_number'].replace('[^0-9]','',regex = True)
#df_no_dnb['odi_number']
```

```python
df_no_dnb['odi_number'] = df_no_dnb['odi_number'].str.replace('[^0-9]','',regex = True).astype('int')
df_no_dnb['odi_number']
```

```python
df_no_dnb.isna().sum()
```

2.converting to int

```python
cols = ['runs_scored','balls_faced','fours','sixes']
#df_no_dnb[cols] = df_no_dnb[cols].apply(lambda x: pd.to_numeric(x,errors = 'coerce'))
df_no_dnb[cols] = df_no_dnb[cols].astype('int')
```

3.converting date column to datetime object

```python
df_no_dnb['date'] = pd.to_datetime(df_no_dnb['date'],format = '%d/%m/%Y')
```

4.converting strike rate to float

```python
df_no_dnb['strike_rate'] = df_no_dnb['strike_rate'].astype('float')
```

5.deriving a new column from score where dhoni is not out

```python
df_no_dnb['not_out'] = np.where(df_no_dnb['score'].str.contains('*',regex = False),True,False)
```

```python
df_no_dnb.head()
```

6.converting score column to int

```python
df_no_dnb['score'] = df_no_dnb['score'].str.replace('[^0-9]','',regex = True).astype('int')
df_no_dnb['score']
```

7.removing v from start of opposition column

```python
df_no_dnb['opposition'] = df_no_dnb['opposition'].str.replace('^[vV]*','',regex = True).str.strip()
df_no_dnb['opposition']
```
