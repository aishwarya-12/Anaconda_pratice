Merging and Joining

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'score': [85, 90, 95, 80]
})
df1
```

```python
df2 = pd.DataFrame({
    'id': [3, 4, 5, 6],
    'name': ['Charlie', 'David', 'Eve', 'Frank'],
    'score': [100, 88, 92, 75]
})
df2
```

merge()

```python
pd.merge(df1,df2, on = 'id')
```

```python
pd.merge(df1,df2, on = 'id', suffixes=('_df1','_df2'))
```

```python
pd.merge(df1,df2, on = 'id', how = 'left',suffixes=('_df1','_df2'))
```

```python
pd.merge(df1,df2, on = ['id','name'],how = 'left',suffixes=('_df1','_df2'))     #joining on two columns
```

```python
df3 = pd.DataFrame({
    '_id': [3, 4, 5, 6],
    'name': ['Charlie', 'David', 'Eve', 'Frank'],
    'score': [100, 88, 92, 75]
})
df3
```

```python
pd.merge(df1,df3, left_on='id', right_on = '_id',suffixes=('_df1','_df3'))        #when key columns of data frames have different names
```

join()  - use it when you want to join on index

```python
df1.join(df2,lsuffix='_df1',rsuffix='_df2',on = 'id')
```

```python
df1.set_index('id',inplace =True)
df2.set_index('id',inplace =True)
```

```python
df1.join(df2,lsuffix='_df1',rsuffix='_df2')      #joining on index and by default its left join
```

```python
df1.join(df2,lsuffix='_df1',rsuffix='_df2', how='inner')       
```

concat()

```python
df1 = pd.DataFrame({
    'emp_id': [101, 102, 103, 104],
    'name': ['John', 'Jane', 'Mike', 'Sara'],
    'department': ['HR', 'Finance', 'IT', 'Marketing']
})
df1
```

```python
df2 = pd.DataFrame({
    'employee_id': [102, 103, 104, 105],
    'salary': [70000, 80000, 75000, 90000],
    'bonus': [5000, 6000, 5500, 7000]
})
df2
```

```python
df_concat = pd.concat([df1,df2],ignore_index = False)      #by deafult its row wise
df_concat
```

```python
new_row = {'emp_id' : 105, 'name' : 'Tom', 'department' : 'IT'}
pd.concat([df1,pd.DataFrame([new_row])],ignore_index=False)
```

```python
new_row = {'emp_id' : 105, 'name' : 'Tom', 'department' : 'IT'}
pd.concat([df1,pd.DataFrame([new_row])],ignore_index=True,sort = True)
```

```python
df_concat = pd.concat([df1,df2],axis =1)          #column wise
df_concat
```
